```python
import numpy as np
import matplotlib.pyplot as plt

# Given parameters
mu = 1      # Mean of the true distribution
sigma2 = 9  # Variance of the true distribution
n = 10      # Sample size

# Create a range of lambda values to plot
lambda_values = np.linspace(0, 10, 100)

# Bias^2 = (lambda * mu / (1 + lambda))^2
bias_squared = (lambda_values * mu / (1 + lambda_values)) ** 2

# Variance = sigma^2 / (n * (1 + lambda)^2)
variance = sigma2 / (n * (1 + lambda_values) ** 2)

# Expected squared error = Bias^2 + Variance
expected_squared_error = bias_squared + variance

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(lambda_values, expected_squared_error, label='Expected Squared Error', lw=2)
plt.plot(lambda_values, bias_squared, label='Bias^2', linestyle='--', lw=2)
plt.plot(lambda_values, variance, label='Variance', linestyle='-.', lw=2)

# Add labels and title
plt.xlabel('Lambda', fontsize=14)

# Add legend
plt.legend(fontsize=12)
plt.savefig('bias_variance.pdf')

# Show grid
#plt.grid(True)

# Display the plot
plt.show()
```
